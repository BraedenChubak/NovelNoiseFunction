```python
%pip install pandas seaborn matplotlib -q
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# Each sample is a 1000x1000 image, chainFreq = 8, jaggedness = 0.25, scrambled = false
df = pd.DataFrame()
for i in range(50):
    filename = "50samples/sample" + str(i) + ".txt"
    current = pd.read_csv(filename, header=None)
    df = pd.concat([df, current], ignore_index=True)
data = df[0]
data.info()
data
```

```python
plt.rcParams.update({'font.size': 12})
plt.figure(figsize=(8, 6))
plt.hist(data, bins=150, density=True, label="Value Distribution", color='darkslategray')
samples = np.random.normal(size=3000, loc=0.5, scale=0.11)
mu, std = stats.norm.fit(samples)
x = np.linspace(0, 1, 100)
y = stats.norm.pdf(x, mu, std)
plt.plot(x, y, 'r', linewidth=2, label='Gaussian Curve')
plt.title("Distribution of Noise Values")
plt.legend()
```
