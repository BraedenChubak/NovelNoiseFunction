```python
%pip install pandas seaborn matplotlib -q
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
df = pd.DataFrame()
for i in range(1):
    filename = "100samples/sample" + str(i) + ".txt"
    current = pd.read_csv(filename, header=None)
    df = pd.concat([df, current], ignore_index=True)
data = df[0]
data.info()
data
# s1 = pd.read_csv("100samples/.txt", header=None)
# s1 = s1[0]
# s1.info()
# s1
```

```python
plt.hist(data, bins=39, density=True, label="100 Samples", color='slategray')
samples = np.random.normal(size=3000, loc=0.4, scale=0.15)
mu, std = stats.norm.fit(samples)
x = np.linspace(0, 1, 100)
y = stats.norm.pdf(x, mu, std)
plt.plot(x, y, 'r', linewidth=2, label='Normal Dist. Centered at 0.4')
plt.legend()
```
